1. Relationship between demografic details and diabetes

```python
def categorize_age(age):
    if age < 10:
        return "0-9"
    elif age < 20:
        return "10-19"
    elif age < 30:
        return "20-29"
    elif age < 40:
        return "30-39"
    elif age < 50:
        return "40-49"
    elif age < 60:
        return "50-59"
    elif age < 70:
        return "60-69"
    elif age < 80:
        return "70-79"
    else:
        return "80+"
    
ds['Age_Category'] = ds['Age'].apply(categorize_age)
```

```python
diagnosis_by_age = ds.groupby(['Age_Category', 'Diagnosis']).size().unstack(fill_value=1)

diagnosis_by_age.columns = ['Healthy', 'Ill']
diagnosis_by_age.plot(kind='bar', stacked=False)
```

```python
diagnosis_by_gender = ds.groupby(['Gender', 'Diagnosis']).size().unstack(fill_value=0)

diagnosis_by_gender.columns = ['Healthy', 'Ill']
diagnosis_by_gender = diagnosis_by_gender.rename(index={0: 'Male', 1: 'Female'})
diagnosis_by_gender.plot(kind='bar', stacked=False)

diagnosis_by_ethnicity = ds.groupby(['Ethnicity', 'Diagnosis']).size().unstack(fill_value=1)
diagnosis_by_ethnicity.columns = ['Healthy', 'Ill']
diagnosis_by_ethnicity = diagnosis_by_ethnicity.rename(index={0: 'Caucasian', 1: 'African American', 2: 'Asian', 3: 'Other'})
diagnosis_by_ethnicity.plot(kind='bar', stacked=False)

diagnosis_by_so_eco_status = ds.groupby(['SocioeconomicStatus', 'Diagnosis']).size().unstack(fill_value=1)
diagnosis_by_so_eco_status.columns = ['Healthy', 'Ill']
diagnosis_by_so_eco_status = diagnosis_by_so_eco_status.rename(index={0: 'Low', 1: 'Middle', 2: 'High'})
diagnosis_by_so_eco_status.plot(kind='bar', stacked=False)

diagnosis_by_education = ds.groupby(['EducationLevel', 'Diagnosis']).size().unstack(fill_value=1)
diagnosis_by_education.columns = ['Healthy', 'Ill']
diagnosis_by_education = diagnosis_by_education.rename(index={0: 'None', 1: 'High School', 2: "Bachelor's", 3: 'Higher'})
diagnosis_by_education.plot(kind='bar', stacked=False)
```

```python
lifestyle_factors = ['BMI', 'AlcoholConsumption', 'PhysicalActivity', 'DietQuality', 'SleepQuality']

plt.figure(figsize=(12, 8))

for i, factor in enumerate(lifestyle_factors, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x='Diagnosis', y=factor, data=ds, palette='Set2')
    plt.title(factor)
    plt.xlabel('Diagnosis (0: No Diabetes, 1: Diabetes)')

plt.tight_layout()
plt.show()
```
